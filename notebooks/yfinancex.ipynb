```python
data_df.columns = ["OPEN", "HIGH", "LOW", "CLOSE", "Adj Close", "Volume"]
data_df
```

```python
# Assign `Adj Close` to `daily_close`
daily_close = data_df[['Adj Close']]
daily_close
```

```python
# Daily returns
daily_pct_change = daily_close.pct_change()
daily_pct_change.fillna(0, inplace=True)
```

```python
daily_pct_change
```

```python
# Average True Range
```

```python
# Define the minumum of periods to consider 
min_periods = 2
```

```python
V = np.sqrt(min_periods) 
V
```

```python
# Calculate the volatility
vol = (daily_pct_change.rolling(min_periods).std() * np.sqrt(min_periods))
vol.columns = ["VOLATILITY"]
vol.fillna(0, inplace=True)
vol
```

```python
# Plot the volatility
vol.plot(figsize=(10, 8))

# Show the plot
plt.show()
```

```python
out = zm.info.get('sharesOutstanding')
```

```python
daily_turnover = data_df[['Volume']]
daily_turnover
```

```python
turnover = (daily_turnover/out) 
turnover.columns = ["TURNOVER"]
turnover
```

```python
data_df["TURNOVER"] = turnover["TURNOVER"]
data_df["VOLATILITY"] = vol["VOLATILITY"]
data_df = data_df.drop(columns = ["Adj Close", "Volume"])
```

```python
df = data_df.reset_index()
df
```

```python
df = df.rename(columns = {"Date": "TIMESTAMP"})
```

```python
new_df = df.round({"OPEN": 2, "HIGH": 2, "LOW": 2, "CLOSE": 2})
new_df
```

```python
TIMESTAMP = new_df["TIMESTAMP"].to_list()
OPEN = new_df["OPEN"].to_list()
HIGH = new_df["HIGH"].to_list()
LOW = new_df["LOW"].to_list()
CLOSE = new_df["CLOSE"].to_list()
TURNOVER = new_df["TURNOVER"].to_list()
VOLATILITY = new_df["VOLATILITY"].to_list()
vol_dict = {"TIMESTAMP": TIMESTAMP, "OPEN": OPEN, "HIGH": HIGH, "LOW ": LOW , "CLOSE": CLOSE, "TURNOVER": TURNOVER, "VOLATILITY": VOLATILITY,}
vol_dict
```

```python
yfinancing_collection.objects.update_one({}, {"$set": vol_dict}, upsert= True)
```
